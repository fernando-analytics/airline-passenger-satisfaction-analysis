```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
print(os.listdir('/'))
```

```python
df = pd.read_csv('/train.csv')

print(df.head())
```

```python
print(df.columns)
```

```python
print(df.info())
```

```python
print(df.head())
```

```python
# -----------------------------
# Satisfaction Distribution
# -----------------------------

satisfaction = df["satisfaction"].value_counts()

plt.figure(figsize=(8,5))

plt.bar(
    satisfaction.index,
    satisfaction.values
)

plt.title("Passenger Satisfaction Distribution")
plt.xlabel("Satisfaction")
plt.ylabel("Number of Passengers")

plt.xticks(rotation=15)

plt.show()


# -----------------------------
# Travel Class Distribution
# -----------------------------

travel_class = df["Class"].value_counts()

plt.figure(figsize=(8,5))

plt.pie(
    travel_class.values,
    labels=travel_class.index,
    autopct='%1.1f%%'
)

plt.title("Passenger Distribution by Travel Class")

plt.show()


# -----------------------------
# Departure Delay Analysis
# -----------------------------

delay = df.groupby(
    "satisfaction"
)["Departure Delay in Minutes"].mean()

plt.figure(figsize=(8,5))

plt.bar(
    delay.index,
    delay.values
)

plt.title("Average Departure Delay by Satisfaction")

plt.xlabel("Satisfaction")
plt.ylabel("Average Delay (Minutes)")

plt.xticks(rotation=15)

plt.show()


# -----------------------------
# Seat Comfort Analysis
# -----------------------------

comfort = df.groupby(
    "satisfaction"
)["Seat comfort"].mean()

plt.figure(figsize=(8,5))

plt.bar(
    comfort.index,
    comfort.values
)

plt.title("Seat Comfort Score by Satisfaction")

plt.xlabel("Satisfaction")
plt.ylabel("Average Comfort Score")

plt.xticks(rotation=15)

plt.show()


# -----------------------------
# Customer Type Distribution
# -----------------------------

customer = df["Customer Type"].value_counts()

plt.figure(figsize=(8,5))

plt.bar(
    customer.index,
    customer.values
)

plt.title("Customer Type Distribution")

plt.xlabel("Customer Type")
plt.ylabel("Passengers")

plt.xticks(rotation=15)

plt.show()
```

```python
print("BUSINESS INSIGHTS")
print("-------------------------------")

print("1. Passenger satisfaction varies significantly across customer groups.")
print("2. Flight delays appear to negatively impact overall satisfaction.")
print("3. Higher seat comfort scores are associated with greater passenger satisfaction.")
print("4. Travel class may influence the overall customer experience.")
print("5. Customer type patterns may reveal differences in loyalty and satisfaction.")
```
